```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

cat_columns = [column for column in df.columns if df[column].dtype == 'O']
num_columns = len(cat_columns)

num_rows = (num_columns // 2) + (num_columns % 2)
num_cols = 2

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))

axes = axes.flatten()


for i, column in enumerate(cat_columns):
    cross_tab = pd.crosstab(df[column], df['Attrition'])
    ax = cross_tab.plot(kind='bar', stacked=True, color=['#4CAF50', '#FF5733'], ax=axes[i])   
    ax.set_title(f"'{column}'")
    ax.set_ylabel("Count")
    ax.set_xlabel(column)

plt.tight_layout()
plt.show()
```

```python
columns_to_drop = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours']
df = df.drop(columns=columns_to_drop, axis=1)
```

```python
sns.set(style="whitegrid")

numeric_columns = df.select_dtypes(include='number').columns

grouped_means = df.groupby('Attrition').mean(numeric_only=True)

grouped_means = grouped_means.T

grouped_means['Difference'] = grouped_means['No'] - grouped_means['Yes']

grouped_means
```

```python
from scipy.stats import ttest_ind

p_values = {}

pd.options.display.float_format = '{:.4f}'.format

for column in numeric_columns:
    t_stat, p_value = ttest_ind(df[df['Attrition'] == 'No'][column], df[df['Attrition'] == 'Yes'][column])
    p_values[column] = p_value

p_values_df = pd.DataFrame(list(p_values.items()), columns=['Variable', 'P-Value'])

p_values_df['Significance Level'] = p_values_df['P-Value'].apply(lambda p: '***' if p < 0.001 else '**' if p < 0.01 else '*' if p < 0.05 else '')

p_values_df
```

```python
more_columns_to_drop = ['Education', 'HourlyRate', 'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'YearsSinceLastPromotion']
log_df = df.drop(columns=more_columns_to_drop, axis=1)
log_df.head(1)
```

```python
log_df['Attrition'] = log_df['Attrition'].replace({'Yes': 1, 'No': 0})
```

```python
import statsmodels.formula.api as smf
logit = smf.logit('Attrition ~Age + C(BusinessTravel) + DailyRate + C(Department) + DistanceFromHome + C(EducationField) + EnvironmentSatisfaction + C(Gender) + JobInvolvement + C(JobRole) + JobLevel + JobSatisfaction + MonthlyIncome + C(MaritalStatus) + C(OverTime) + StockOptionLevel + TotalWorkingYears + TrainingTimesLastYear + WorkLifeBalance + YearsAtCompany + YearsInCurrentRole + YearsWithCurrManager', data=log_df).fit(maxiter=10000)
print(logit.summary())
```
